```python
import numpy as np
import pandas as pd

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

train_data = pd.read_csv('data/train.csv')
test_data = pd.read_csv('data/test.csv')
print(train_data.head())
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 1번 방법
# 기존 코드에서 'age', 'Fare' 추가

train_d = train_data.copy()
train_d = train_data.dropna(subset=['Age'])

test_d = test_data.copy()
train_d = train_data.dropna(subset=['Age'])
test_d['Fare'] = test_d['Fare'].fillna(test_d['Fare'].mean())

from sklearn.ensemble import RandomForestClassifier
features = ['Pclass', 'Sex', 'Age', 'Fare' ,'SibSp', 'Parch']

train_input = pd.get_dummies(train_d[features])
test_input = pd.get_dummies(test_d[features])

train_target = train_d['Survived']

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
rfc.fit(train_input, train_target)

predictions = rfc.predict(test_input)

result = {
    'PassengerId': test_d['PassengerId'],
    'Survived': predictions,
}

output = pd.DataFrame(result)

output.to_csv('data/my_submission.csv', index=False)
```

```python
# 2번
import pandas as pd
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression

# 데이터 복사
train_copy = train_data.copy()
test_copy = test_data.copy()

test_copy['Fare'] = test_copy['Fare'].fillna(test_copy['Fare'].mean())

# 타겟 변수 재설정 (결측치 처리 후)
train_target = train_copy['Survived']

# 특성 선택
features = ['Pclass', 'Sex', 'Fare', 'SibSp', 'Parch']

# 원-핫 인코딩
train_input = pd.get_dummies(train_copy[features])
test_input = pd.get_dummies(test_copy[features])

# 정규화
ss = StandardScaler()
train_scaled = ss.fit_transform(train_input)
test_scaled = ss.transform(test_input)

# 다항식 특성 생성
poly = PolynomialFeatures(degree=7)
train_poly = poly.fit_transform(train_scaled)  # fit_transform 사용
test_poly = poly.transform(test_scaled)

# 선형 회귀 모델 학습
lr = LinearRegression()
lr.fit(train_poly, train_target)  # 샘플 수가 일치해야 함

# 예측
predictions = lr.predict(test_poly)

# 예측 결과를 정수형으로 변환
predictions = (predictions > 0.5).astype(int)

# 결과 저장
result = {
    'PassengerId': test_data['PassengerId'],
    'Survived': predictions,
}

output = pd.DataFrame(result)

# CSV로 저장
output.to_csv('data/my_submission.csv', index=False)
```
